```python
# import required modules
from bs4 import BeautifulSoup
import requests
from IPython.display import Image, display


 
# get URL
url_base = 'https://es.wikipedia.org'
#page = requests.get(url_base + "/wiki/Anexo:Banderas_de_Estados_soberanos")
#page = requests.get(url_base + "/wiki/Anexo:Banderas_de_Estados_con_reconocimiento_limitado")
page = requests.get(url_base + "/wiki/Anexo:Banderas_de_territorios_dependientes")
 
# scrape webpage
soup = BeautifulSoup(page.content, 'html.parser')
 
list(soup.children)

links = soup.select('.image')

#print(links[0])
#print('\n\n')

i = 1
for link in links:
    pais = link['title']
    url_thumb = 'https:' + link.img['src']
    url_archivo = url_base + link['href']
    imagenes = [imagen.split(' ')[0] for imagen in link.img['srcset'].split(', ')]
    url_bandera = 'https:' + imagenes[1]
    url_bandera = url_bandera.split('.svg')[0].replace('/thumb', '') + '.svg'
    
    nombre_archivo_svg = pais.lower().replace(' ', '_') + '.svg'
    
    print(pais)
    #print(url_bandera)
    #print(nombre_archivo_svg)
    #display(Image(url_thumb))    
    response = requests.get(url_bandera)
    open("../../data/banderas/" + nombre_archivo_svg, "wb").write(response.content)    
    print('\n')
    i += 1
```
