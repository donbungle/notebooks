```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from pathlib import Path  
```

```python
filepath = Path('../../data/peliculas.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
```

```python
i = 1
j = 1
titles = []
years = []
votes = []
indices = []
while True:
    URL = 'https://ww1.cuevana3.me/peliculas/page/{0}'.format(i)
    page = requests.get(URL)
    status = page.status_code
    print('{2} {1} and {0}'.format(i, status, URL))
    if status != 200:
        break
    
    soup = BeautifulSoup(page.content, "html.parser")
    results = soup.find(id="aa-wp")
    lista = results.find_all("li", class_="xxx")    
    print('Largo lista: {0}'.format(len(lista)))
    if len(lista) <= 0:
        break
        
    titles = []
    years = []
    votes = []
    indices = []
    descs = []
    genres = []
    
    for elemento in lista:
        title_element = elemento.find("h2", class_="Title")
        year_element = elemento.find("span", class_="Year")
        vote_element = elemento.find("span", class_="Vote AAIco-star")
        qlty_element = elemento.find("span", class_="Qlty")
        desc_element = elemento.find("div", class_="Description")
        genre_element = elemento.find("p", class_="Genre AAIco-movie_creation")
        
        #print(title_element.text)
        indices.append(j)
        titles.append(title_element.text)
        years.append(year_element.text)
        votes.append(vote_element.text)
        descs.append(desc_element.text)
        genres.append(genre_element.text)
        j += 1
        
    data = {
        "id": indices,
        "title": titles,
        "year": years,
        "vote": votes,
        "desc": descs,
        "genre": genres,
    }
    df_peliculas = pd.DataFrame(data)
    df_peliculas.to_csv(filepath, mode='a', index=False, header=False)
    i += 1
```
